##### 6.Make an OrderedDict called fancy from the same pairs listed in 5 and print it. Did it print in the same order as plain?


```python
# A Python program to demonstrate working of OrderedDict
from collections import OrderedDict
```

```python
from collections import OrderedDict

key_value = (('a', 1), ('b', 2), ('c', 3))

print('dict:')
for key, value in dict(key_value).items():
    print(key, value)
    
print('ordered dict:')
for key, value in OrderedDict(key_value).items():
    print(key, value)

```

##### 7. Make a default dictionary called dict_of_lists and pass it the argument list. Make the list dict_of_lists[&#39;a&#39;] and append the value &#39;something for a&#39; to it in one assignment. Print dict_of_lists[&#39;a&#39;].

```python
from collections import defaultdict
      
# Defining the dict
dict_of_lists = defaultdict(list)
print(type(dict_of_lists))
dict_of_lists['a'].append('something for a')
print(dict_of_lists['a'])

```
